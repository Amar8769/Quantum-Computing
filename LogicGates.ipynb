```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np
```

#1. Quantum Simulation of Classical Logic Gates

XOR Gate (CNOT)

```python
backend = AerSimulator()

qc_040 = QuantumCircuit(3)
qc_040.h(0)
qc_040.h(1)

qc_040.cx(0, 2)
qc_040.cx(1, 2)
qc_040.measure_all()

qc_040.draw('mpl')
```

```python
result = backend.run(qc_040).result().get_counts()
plot_histogram(result, title = "Amar Harshul Vaid")
```

AND Gate (Toffoli-based)

```python
backend = AerSimulator()

qc_040 = QuantumCircuit(3)
qc_040.h(0)
qc_040.h(1)
qc_040.ccx(0, 1, 2)
qc_040.measure_all()

qc_040.draw('mpl')
```

```python
result = backend.run(qc_040).result().get_counts()
plot_histogram(result, title = "Amar Harshul Vaid")
```

OR Gate (Using De Morgan’s Law)

```python
backend = AerSimulator()

qc_040 = QuantumCircuit(3)

qc_040.h(0)
qc_040.h(1)
qc_040.cx(0, 2)
qc_040.cx(1, 2)
qc_040.ccx(0, 1, 2)

qc_040.measure_all()

qc_040.draw('mpl')
```

```python
result = backend.run(qc_040).result().get_counts()
plot_histogram(result, title = "Amar Harshul Vaid")
```

#2. Quantum Half Adder

```python
backend = AerSimulator()

qc_040 = QuantumCircuit(4)
qc_040.h(0)
qc_040.h(1)

qc_040.cx(0, 2)
qc_040.cx(1, 2)

qc_040.ccx(0, 1, 3)

qc_040.measure_all()

qc_040.draw('mpl')
```

```python
result = backend.run(qc_040).result().get_counts()
plot_histogram(result, title = "Amar Harshul Vaid")
```

#3. Quantum Full Adder

```python
backend = AerSimulator()

qc_040 = QuantumCircuit(5)
qc_040.h(0)
qc_040.h(1)
qc_040.h(2)

# SUM = XOR of all three
qc_040.cx(0, 3)
qc_040.cx(1, 3)
qc_040.cx(2, 3)

# CARRY = majority logic
qc_040.ccx(0, 1, 4)
qc_040.ccx(1, 2, 4)
qc_040.ccx(0, 2, 4)

qc_040.measure_all()

qc_040.draw('mpl')
```

```python
result = backend.run(qc_040).result().get_counts()
plot_histogram(result, title = "Amar Harshul Vaid")
```
