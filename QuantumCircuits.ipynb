```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_histogram
from qiskit_aer import AerSimulator
import math
```

```python
# Circuit 1 – Basic Measurement of |0>
QR_040 = QuantumRegister(1, 'Q')
CR_040 = ClassicalRegister(1, 'C')
QC_040 = QuantumCircuit(QR_040, CR_040)

QC_040.measure_all()
QC_040.draw('mpl')
```

```python
sim = AerSimulator()
result = sim.run(QC_040).result().get_counts()
plot_histogram(result, title="Amar Harshul Vaid")
```

```python
# Circuit 2 – Hadamard Superposition
QR_040 = QuantumRegister(1, 'Q')
CR_040 = ClassicalRegister(1, 'C')
QC_040 = QuantumCircuit(QR_040, CR_040)

QC_040.h(0)
QC_040.barrier()
QC_040.measure_all()
QC_040.draw('mpl')
```

```python
sim = AerSimulator()
counts = sim.run(QC_040).result().get_counts()
plot_histogram(counts, title="Amar Harshul Vaid")
```
