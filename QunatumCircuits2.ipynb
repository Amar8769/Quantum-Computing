```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
```

```python
sim = AerSimulator()
```

```python
qcI = QuantumCircuit(1, 1)
qcI.id(0)
qcI.measure(0, 0)
qcI.draw('mpl')
```

```python
resultI = sim.run(qcI).result()
countsI = resultI.get_counts()
plot_histogram(countsI, title="Identity Gate - Amar Harshul Vaid")
```

```python
stateI = Statevector.from_instruction(qcI.remove_final_measurements(inplace=False))
plot_bloch_multivector(stateI)
```
