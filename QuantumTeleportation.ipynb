```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import matplotlib.pyplot as plt
import numpy as np
```

```python
QC_040=QuantumCircuit(3,3)
QC_040.draw('mpl')
```

```python
QC_040.initialize([np.sqrt(3)/2, 1/2], 0)
```

```python
QC_040.draw('mpl')
```

```python
state0 = Statevector(QC_040)
```

```python
state0.draw(output='latex')
```

```python
plot_bloch_multivector(state0)
```

```python
QC_040.h(1)
QC_040.cx(1,2)
```

```python
QC_040.draw('mpl')
```

```python
state1 = Statevector(QC_040)
```

```python
state1.draw(output='latex')
```

```python
plot_bloch_multivector(state1)
```

```python
QC_040.cx(0,1)
QC_040.h(0)
```

```python
QC_040.draw('mpl')
```

```python
state2 = Statevector(QC_040)
```

```python
state2.draw(output='latex')
```

```python
plot_bloch_multivector(state2)
```

```python
QC_040.measure([0,1],[0,1])
```

```python
Simulator=AerSimulator()
```

```python
job=Simulator.run(QC_040)
result=job.result()
counts=result.get_counts()
```

```python
plot_histogram(counts,title='Amar Harshul Vaid')
```

```python
with QC_040.if_test( (QC_040.clbits[1], 1) ):
  QC_040.x(2)
```

```python
with QC_040.if_test( (QC_040.clbits[0], 1) ):
  QC_040.z(2)
```

```python
QC_040.measure(2,2)
```

```python
QC_040.draw('mpl')
```

```python
job=AerSimulator().run(QC_040)
result=job.result()
counts=result.get_counts()
```

```python
plot_histogram(counts,title='Amar Harshul Vaid')
```
