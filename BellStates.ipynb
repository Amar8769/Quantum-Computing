```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit_aer import AerSimulator
from numpy import pi, sqrt
import numpy as np
from matplotlib import pyplot as plt
```

## 1. The Four Bell States

Bell State 1 — |Φ⁺⟩ = (|00⟩ + |11⟩)/√2

```python
QC_040 = QuantumCircuit(2)
QC_040.h(0)
QC_040.cx(0, 1)
QC_040.measure_all()

backend = AerSimulator()
result = backend.run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

```python
plot_histogram(result,title = "Amar Harshul Vaid")
```

Bell State 2 — |Φ⁻⟩ = (|00⟩ − |11⟩)/√2

```python
QC_040 = QuantumCircuit(2)
QC_040.rz(np.pi, 1)     # Z rotation by π on qubit 1
QC_040.h(0)
QC_040.cx(0, 1)
QC_040.measure_all()

result = AerSimulator().run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

```python
plot_histogram(result,title = "Amar Harshul Vaid")
```

Bell State 3 — |Ψ⁺⟩ = (|01⟩ + |10⟩)/√2

```python
QC_040 = QuantumCircuit(2)
QC_040.x(1)
QC_040.h(0)
QC_040.cx(0, 1)
QC_040.measure_all()

result = AerSimulator().run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

```python
plot_histogram(result,title = "Amar Harshul Vaid")
```

Bell State 4 — |Ψ⁻⟩ = (|01⟩ − |10⟩)/√2

```python
QC_040 = QuantumCircuit(2)
QC_040.x(1)
QC_040.rz(np.pi, 0)
QC_040.h(0)
QC_040.cx(0, 1)
QC_040.measure_all()

result = AerSimulator().run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

#2. The SWAP Operation (using 3 CNOTs)

```python
QC_040 = QuantumCircuit(2)
# Initialize q0 = cos(π/3)|0⟩ + sin(π/3)|1⟩, q1 = |1⟩
QC_040.ry(2*np.pi/3, 0)
QC_040.x(1)

# Show Bloch spheres before SWAP
State_before = Statevector(QC_040)
plot_bloch_multivector(State_before)
```

```python
# SWAP using 3 CNOTs
QC_040.cx(0, 1)
QC_040.cx(1, 0)
QC_040.cx(0, 1)

# After SWAP
State_after = Statevector(QC_040)
plot_bloch_multivector(State_after)
```

#3. 3-Qubit Partial Entanglement and Measurement

```python
QC_040 = QuantumCircuit(3)
QC_040.h(2)
QC_040.cx(2, 1)
QC_040.cx(1, 0)
QC_040.h(0)

State = Statevector(QC_040)
plot_bloch_multivector(State)
```

```python
QC_040.measure_all()
result = AerSimulator().run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

```python
plot_histogram(result,title = "Amar Harshul Vaid")
```

#5. Demonstrating the Toffoli (CCNOT) Gate

Case 1 — Demonstrate Conditional Flip

```python
QC_040 = QuantumCircuit(3)
QC_040.h(0)
QC_040.h(1)
QC_040.ccx(0, 1, 2)
QC_040.measure_all()

result = AerSimulator().run(QC_040).result().get_counts()

QC_040.draw('mpl')
```

```python
plot_histogram(result,title = "Amar Harshul Vaid")
```
