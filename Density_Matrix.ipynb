```python
QC_040=QuantumCircuit(2)
QC_040.h(0)
QC_040.cx(0,1)
QC_040.draw('mpl')
```

```python
state = Statevector.from_instruction(QC_040)
print("Statevector =",state)
```

```python
rho_bell_state=DensityMatrix(state)
```

```python
plot_state_city(rho_bell_state,title = "Amar Harshul vaid")
```

```python
#1.
rho_0_040=DensityMatrix.from_label('0')
rho_1_040=DensityMatrix.from_label('1')

rho_040=0.8*rho_0_040 + 0.2*rho_1_040
plot_state_city(rho_040,title = "Amar Harshul vaid")
```

```python
#2.
# |> = (|0> + i|1>)/2
psi = Statevector([1/2**0.5, 1j/2**0.5])

# Compute density matrix
rho_040 = psi.to_operator().data
print(rho_040)
```

```python
plot_state_city(rho_040,title = "Amar Harshul vaid")
```
