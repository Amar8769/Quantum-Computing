```python
!pip install qiskit qiskit-aer pylatexenc

from qiskit.quantum_info import DensityMatrix, Statevector, partial_trace
from qiskit.visualization import plot_state_city
from qiskit import QuantumCircuit
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_purity(rho_040):
    rho_squared = rho_040.data @ rho_040.data
    purity_value = np.trace(rho_squared)
    return purity_value
```

```python
rho_pure_040 = DensityMatrix.from_label('0')

purity_calc_040 = calculate_purity(rho_pure_040)
purity_builtin_040 = rho_pure_040.purity()

print(f"Calculated Purity = {purity_calc_040:.6f}")
print(f"Built-in Purity = {purity_builtin_040:.6f}")
```

```python
plot_state_city(rho_pure_040, title="Amar Harshul Vaid – Pure |0⟩ State")
```

```python
rho_0_040 = DensityMatrix.from_label('0')
rho_1_040 = DensityMatrix.from_label('1')
rho_mixed_040 = 0.6 * rho_0_040 + 0.4 * rho_1_040

purity_calc_040 = calculate_purity(rho_mixed_040)
purity_builtin_040 = rho_mixed_040.purity()

print(f"Calculated Purity = {purity_calc_040:.6f}")
print(f"Built-in Purity = {purity_builtin_040:.6f}")
```

```python
def entanglement_create_mixed_states_040():
    QC_040 = QuantumCircuit(2)
    QC_040.h(0)
    QC_040.cx(0, 1)

    bell_state_040 = Statevector.from_instruction(QC_040)
    rho_bell_040 = DensityMatrix(bell_state_040)

    print(f"Bell State purity : {rho_bell_040.purity():.3f}")

    # Trace out qubit 1
    rho_qubit0_040 = partial_trace(rho_bell_040, [1])
    print(f"Qubit 0 purity after partial trace : {rho_qubit0_040.purity():.3f}")

    # Trace out qubit 0
    rho_qubit1_040 = partial_trace(rho_bell_040, [0])
    print(f"Qubit 1 purity after partial trace : {rho_qubit1_040.purity():.3f}")
```

```python
entanglement_create_mixed_states_040()
```

```python
def print_purities_040(name_040, state_040):
    print(f"\n{name_040}")
    rho_full_040 = DensityMatrix(state_040)

    print(f"a) Full system purity : {rho_full_040.purity():.3f}")

    rho_qubit0_040 = partial_trace(rho_full_040, [1])
    print(f"b) Qubit 0 purity after partial trace : {rho_qubit0_040.purity():.3f}")

    rho_qubit1_040 = partial_trace(rho_full_040, [0])
    print(f"c) Qubit 1 purity after partial trace : {rho_qubit1_040.purity():.3f}")
```

```python
def purity_theta_040(theta_040):
    QC_040 = QuantumCircuit(2)
    QC_040.ry(2 * theta_040, 0)
    QC_040.cx(0, 1)

    state_040 = Statevector.from_instruction(QC_040)
    rho_040 = DensityMatrix(state_040)
    rho_A_040 = partial_trace(rho_040, [1])

    purity_val_040 = np.real(np.trace(rho_A_040.data @ rho_A_040.data))
    print(f"\nθ = {theta_040:.2f} radians → Purity of Qubit 0: {purity_val_040:.3f}")
    return purity_val_040, rho_040, rho_A_040
```

```python
theta_040 = float(input("Enter θ (in radians): "))
purity_val_040, rho_full_040, rho_A_040 = purity_theta_040(theta_040)

plot_state_city(rho_full_040, title="Amar Harshul Vaid – Full 2-Qubit State")
plot_state_city(rho_A_040, title="Amar Harshul Vaid – Reduced Qubit 0 (Partial Trace)")
```
